```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Explore the distribution of variables
variables = ['movie_name', 'rating_target_audience', 'genre', 'release_year_and_country', 'release_date','score','votes','director','writer','star','country','budget','gross','company','runtime']
for variable in variables:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=data, x=variable, hue='rating_target_audience', kde=True)
    plt.title(f'Distribution of {variable} by rating_target_audience')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='rating_target_audience', y='score')
plt.title('score distribution by rating_target_audience')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='rating_target_audience', y='runtime')
plt.title('runtime Distribution by rating_target_audience')
plt.show()
```

```python
#Dealing with missing values
data.isnull().sum()
```

```python
# Check for missing data
data.info()
```

```python
# Fill missing values in numeric columns with mean
data['score'].fillna(data['score'].mean(), inplace=True)
```

```python
data['votes'].fillna(data['votes'].mean(), inplace=True)
```

```python
data['budget'].fillna(data['budget'].mean(), inplace=True)
```

```python
data['gross'].fillna(data['gross'].mean(), inplace=True)
```

```python
data['runtime'].fillna(data['runtime'].mean(), inplace=True)
```

```python
# Fill missing values in a categorical column with mode
data['rating_target_audience'].fillna(data['rating_target_audience'].mode()[0], inplace=True)
data['release_year_and_country'].fillna(data['release_year_and_country'].mode()[0], inplace=True)
data['director'].fillna(data['director'].mode()[0], inplace=True)
data['writer'].fillna(data['writer'].mode()[0], inplace=True)
data['star'].fillna(data['star'].mode()[0], inplace=True)
data['country'].fillna(data['country'].mode()[0], inplace=True)
data['company'].fillna(data['company'].mode()[0], inplace=True)
```
